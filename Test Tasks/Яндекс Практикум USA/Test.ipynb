**Предметная область:** онлайн платформа с несколькими обучающими программами (профессиями). Программы состоят из уроков. Студенты проходят уроки на платформе


**Даны таблицы:**
finished_lesson_test - таблица со всеми прохождениями уроков. Одна строчка - одно прохождение конкретным студентом конкретного урока конкретной профессии.

**Поля:**
* date_created - время завершения урока студентом
* id - уникальный айди записи в таблице
* lesson_id - айди урока как единицы контента. Внешний ключ для связи с таблицей lesson_index_test
* user_id - айди прошедшего урок ученика
* lesson_index_test - таблица со всеми уроками, доступными на платформе.


**Поля:**
* lesson_id - ключ таблицы
* lesson_name - название урока
* profession_id - айди профессии урока
* profession_name - название профессии


Каждый урок может быть пройден студентом только один раз. Студент может проходить уроки разных профессий.


**Задача:**

После раскатки последнего обновления платформы в логах начал появляться баг: иногда некоторые уроки, согласно записям, проходятся подозрительно “быстро”. Проблемными считаем кейсы, когда между прохождением двух уроков программы студентом прошло не более пяти секунд. Нужно написать SQL-запрос, который, опираясь на таблицы finished_lesson_test и lesson_index_test, выдаст выгрузку проблемных кейсов вида:

|delta_seconds|lession_datetime|lesson_id|next_lesson_datetime|profession_name|user_id|
|:------------|:---------------|:--------|:-------------------|:--------------|:------|
|0|2020-04-09T17:44:10Z|"21fbfb7c-8b74-40e7-9cd8-9590b7903f26"|2020-04-09T17:44:10Z|"data-analyst"|535641|
|2|2020-04-09T17:44:10Z|"24d355f6-4f88-4c05-bf51-e0e1db42f184"|2020-04-09T17:44:12Z|"data-analyst"|535645|
|4|2020-04-09T17:44:10Z|"3891bd65-5ff9-4355-b886-f5796ce6ba72"|2020-04-09T17:44:14Z|"data-analyst"|535648|

содержащую прохождения уроков, за которыми последовало прохождение следующего урока той же профессии не более чем через 5 секунд (столбец **delta_seconds** в выгрузке - разница в секундах между временем прохождения следующего урока и текущего). (**next_lesson_datetime** в выгрузке - это время прохождения следующего урока).
При этом для начала мы хотим ограничиться одной когортой конкретной профессии. Выгрузка должна содержать данные **только для студентов профессии data-analyst апрельской когорты 2020 года**. В выгрузке не должны учитываться уроки других когорт или уроки данной когорты, но в других профессиях. К апрельской когорте профессии data-analyst 2020 года относятся те студенты, которые прошли свой первый урок в этой профессии в апреле 2020 года.

База: postgreSQL

### Решение <a class='ancchor' id='6'>

По условиям, данный код должен быть отправлен через платформу, на которой проходит тестирование. И отображен тут он должен быть лишь в том случае, когда платформа его не принимает из-за скоросты исполнения.

Данный код принимается платформой и тут отображен скорее для справки.

N.B.: в случае использования шардированного хранилища данных - внутренний запрос нижнего уровня ("отберем только аналитиков данных") имеет смысл сделать в виде CTE и далее использовать СТЕ. Это поможет избежать проблемы "неполной вязки" данных.

````python
SELECT
    delta_seconds delta_seconds,
    lesson_datetime,
    lesson_id,
    next_lesson_date_time,
    profession_name,
    user_id
FROM( 
    -- найдем время в секундах между прохождением двух уроков одним пользователем из апрельской когорты data-analyst
    SELECT
        ROUND(EXTRACT(EPOCH FROM (LAG(date_created) OVER (PARTITION BY user_id, profession_id  ORDER BY date_created DESC) - date_created))) delta_seconds, 
        date_created lesson_datetime,
        lesson_id,
        LAG(date_created) OVER (PARTITION BY user_id, profession_id  ORDER BY date_created DESC) next_lesson_date_time,
        profession_name,
        user_id
    FROM(
        -- найдем пользователей из апрельской когорты
        SELECT
            user_id,
            date_created,
            lesson_id,
            lesson_name,
            profession_id,
            profession_name
        FROM(
            -- найдем "дату рождения" каждого аналитика
            SELECT
                user_id,
                min(date_created) OVER(partition by user_id, profession_name) birth_date,
                date_created,
                lesson_id,
                lesson_name,
                profession_id,
                profession_name
            FROM(
                -- отберем только аналитиков данных
                SELECT
                    t1.user_id,
                    t1.date_created,
                    t1.lesson_id,
                    t2.lesson_name,
                    t2.profession_id,
                    t2.profession_name
                FROM finished_lesson_test t1 left join lesson_index_test t2 USING (lesson_id)
                WHERE t2.profession_name = 'data-analyst'
            )tmp_0
            GROUP BY
                user_id,
                date_created,
                lesson_id,
                lesson_name,
                profession_id,
                profession_name
        )tmp_1
        WHERE tmp_1.birth_date between '2020-04-01 00:00:00-11:00' and '2020-04-30 00:00:00-11:00'
    )tmp_2
)tmp_3
WHERE
    delta_seconds<=5
ORDER BY
    delta_seconds
````

Другими словами: если средний чек в группе В не падает - то выручка точно растет, а если падает, то нужно считать на сколько и может ли рост числа покупок скомпенсировать падение чека.

Проверим сначала первое. Будем идти от простого к сложному и сначала посмотрим, можем ли мы оценить наши результаты t-тестом. Для его использования нужно выполнение ряда условий:
- некоррелируемость признаков
- нормальность распределения признаков
- гомогенность дисперсий признаков

Для проверки на нормальность можно использовать, например тест Шапиро-Вилка. При этом, нужно понимать, что даже отвергание гипотезы о нормальности распределения не дает однозначного запрета на использование критерия Стьюдента потому что доказано, что при больших выборках (где n>30) требование нормальности распределения для t-критерия является необязательным.

При проверрке на гомогенность дисперсии можно использовать критерий Фишера. Тут важно помнить, что в отличае от t-теста, для F-критерия требование нормальности распределений является обязательнымзатем и, в случае нарушения этого требования, можно легко, без регистраций и смс, получить неадекватные результаты. Поэтому, если вдруг наши выборки распределены не нормально, то следует использовать какой-либо непараметрический аналог, например тест Крускала-Виллиаса (он, кстати, хорошо работает если выборки разного размера).

````python
def can_we_use_t_test(sample_1, sample_2, alpha=0.05):
    
    """
    can_we_use_t_test(sample_1, sample_2, alpha=0.05)
    проверяет условия использования t-критерия Стъюдента
    для сравнения средних 2х выборок.
    sample_1 - numpy array
    sample_2 - numpy array
    alpha - уровень значимости
    """

    # проверим распределения на нормальность
    shapiro_1 = stats.shapiro(sample_1).pvalue
    shapiro_2 = stats.shapiro(sample_2).pvalue

    # проверим гомогенность дисперсии
    var_1 = (sample_1 - sample_1.mean())**2
    var_2 = (sample_2 - sample_2.mean())**2

    var_1_shapiro = stats.shapiro(var_1).pvalue
    var_2_shapiro = stats.shapiro(var_2).pvalue

    if var_1_shapiro >= alpha and var_2_shapiro >= alpha:

        F = np.var(sample_1) / np.var(sample_2)
        df_1 = len(sample_1)-1
        df_2 = len(sample_2)-1
        homogeneity_p = 1 - st.f.cdf(F, df_1, df_2)

    else:
        homogeneity_p = stats.kruskal(sample_1, sample_2).pvalue



    if (shapiro_1 >= alpha or (shapiro_1 < alpha and len(sample_1) > 30)) and \
     (shapiro_2>= alpha or (shapiro_2 < alpha and len(sample_2)>30)) and \
     homogeneity_p > alpha:

        print('проверяем гипотезу по Стьюденту')
    else:
        print('следует выбрать другой критерий')
        
        
can_we_use_t_test(orders[orders.group == 'A'].revenue.values, orders[orders.group == 'B'].revenue.values)
````

Отлично. Тест показывает, что мы можем проверять по Стьюденту. Что мы сейчас и сделаем.
