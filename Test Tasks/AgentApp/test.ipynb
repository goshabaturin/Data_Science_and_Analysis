# Тестовое задание

На позицию middle Data scientist в AgentApp. 

Выполнил Георгий Батурин.


Цель задания: понять насколько хорошо соискатель может поставить цели и спланировать работу неболшой команды data science проекта.


## Формулировка

* **Задача команды**

    - построение аналитической модели для выявления нежелательных договоров ОСАГО на основании данных (подготовленных витрин) по договорам, клиентам, убыткам, ТС;
    - оценка построенной модели на бэк-тесте (тестовой исторической выборке);
    - Оценка экономической выгоды от применения аналитической модели с последующей экстраполяцией полученных результатов на всю базу договоров и котировок ОСАГО
<br><br>
* **Ожидаемый результат работы команды**

    - витрина данных с предикторами, сформированными на исходных данных Заказчика, обогащенная производными параметрами, агрегированными с помощью математических функций и с учетом различных временных отрезков;
    - модель, определяющая вероятность того, что договор будет являться убыточным для Заказчика со списком значимых предикторов модели с показателями значимости и показателями точности модели;
    - расчет и оценка экономической выгоды, достигнутой на пилотном проекте
<br><br>

* **В результате выполнения задания требуется описать**

    - Этапы проекта (описание этапов от подготовки данных до достижения результата с указанием задач для группы исполнителей)
    - Критерии достижения целей и приемки результатов проекта (измеримые критерии и их значения, достижение которых подтверждает соответствие результатов проекта (этапа проекта) поставленным целям)
    - Контрольные точки проекта (перечень ключевых события проекта: результаты, которые должны быть получены по состоянию на прогнозные даты)

## Решение

**<u>Основные этапы проекта</u>**
1. Формализация задачи;
2. Организация доступа к данным и их валидация;
3. Подготовка признаков;
4. Анализ признаков;
5. Создание и наполнения витрины данных для предикторов модели;
6. Выбор группы наиболее перспектиыных (с точки зрения цели) алгоритмов; 
7. Создание base-line модели;
8. Создание модели;
9. "Тонкая" настройка параметров модели, повышение точности
10. Оценка экономической и бизнес - выгоды от внедрения в продакшн.
11. Развертывание модели в продакшн;
12. Определение дополнительных данных для улучшения качества;
13. Cбор дополнительных данных;
14. Повторение пунктов 2-12.


**<u>Контрольные и ключевые точки проекта</u>**

Окончание каждого из следующих пунктов можно считать контрольной точкой проекта.  Окончание пунктов 5, 10 и <u>успешное</u> окончание 11 пункта -  можно считать прохождением ключевых контрольных точек проекта. Успешное окончание пункта 11 - точка окончания проекта и переход от стадии "проект" к стадии "продукт".

**<u>Более подробно</u>**

Для простоты повествования пойдем по основным этапам с детализацией по действиям и действующим лицам.

**1.** 

Первоначально, совместно с заказчиком, необходимо сформировать задачу в терминах анализа данных и машинного обучения, формализовать в виде четкого критерия / критериев / взаимодействия критериев определение бизнес термина "нежелательный договор ОСАГО". Необходимо определить метрику/метрики по которым мы будем оценивать работу модели. Если модель вероятностная - необходимо определить пороговую вероятность отнесения объектов к классу.
Основными действующими лицами этой стадии являются: заказчик, лид DS (здесь и далее - data scientist), специалисты DA (здесь и далее - data analyst) и DS выделенные на проект.

**2.** 

Поскольку есть какой-то набор данных от заказчика - нужно сначала получить, а затем - изучить данные. Необходимо понимать:  
- какие данные есть и что они обозначают в реальном мире, при необходимости - как дешифруются в понятные человеку данные (например, "значение "25" в поле region соответствует Приморскому краю и дешифруется согласно правилам дешифровки автомобильных кодов регионов РФ"); 
- валидировать входные данные: разобраться с тем как данные собираются, какие значения могут принимать поля, насколько полно собираются данные (все ли данные полежашие сбору попадают в систему или что-то теряется по пути). Как пример - нужно знать ответ на вопрос "вносятся ли поле "запасное колесо на задней двери" руками в специальную форму или автоматически вносится системой на основании распознования фотографии". От этого зависит, как мы будем обрабатывать пропуски значений. Так, если при оценке авто, поле "запасное колесо на задней двери" - не обязательный, а данные вносятся в базу "руками" то пропуски данных в этом поле может указывать не на нарушение целостности входных данных, а на отсутствие запасного колеса, например, потому что оценивается мотоцикл;
- на всех выше описанных этапах, в случае возникновения вопросов - имеет смысл общаться с заказчиком и, возможно, выставлять тз на доработки витрин.
Основными действующими лицами этой стадии являются: заказчик, специалисты BA (здесь и далее - bussiness analyst), DA, DS, DE (здесь и далее - data engineer) выделенные на проект. 

**3.**

Когда мы понимаем природу данных - самое время приступить к их подготовке: обработать пропуски, кодировать категориальные переменные, привести типы, удалить дубликаты, исправить форматы и тд.
Основными действующими лицами этой стадии являются: специалист DA.

**4.**

Когда данные готовы - можно проводить анализ входных данных: искать закономерности, зависимости, трансформировать пространство признаков (уменьшать/увеличивать/ переводить в другую систему координат и тд), комбинировать несколько признаков в один, разбивать один признак на несколько и тд.
Основными действующими лицами этой стадии являются: специалист DA.

**5.**

После того, как признаки готовы - можно получить первый  ожидаемый результат: " витрина данных с предикторами, сформированными на исходных данных Заказчика, обогащенная производными параметрами, агрегированными с помощью математических функций и с учетом различных временных отрезков". Для этого нужно согласовать (со всеми стейкхолдерами) формата витрины, создать и автоматизировать наполнение витрины.
Основными действующими лицами этой стадии являются: заказчик, лид DS, специалисты DE, DS, DA, BA.

**6.**

Выбор, на основании знания о имеющихся данных, группы наиболее перспективных алгоритмов, с точки зрения теоретически возможной предсказательной силы. Выбор метрик, по которым будут происходить минимизация функции потерь модели. Если необходимо - приведение признаков к единому масштабу и другие подготовки для передачи признаков в модель, выбор метода обучения, валидации и тестирования результатов моделей, разбиение выборки.
Основными действующими лицами этой стадии являются: специалист DS, в зависимости от квалификации специалиста - на этапе выбора наиболее перспективной группы алгоритмов и функции потерь - может подлючаться или не подключаться лид DS.

**7.**

Прототипирование "модели начального уровня", основная цель которой - быть лучше случайности.
Основными действующими лицами этой стадии являются: специалист DS.

**8.**

Создание моделей, проведение соревнования между всеми моделями, включая base-line. Выбор наиболее удачной. При необходимости - создание "мега-модели" (bagging, boosting, stacking) из нескольких наиболее удачных моделей.
Основными действующими лицами этой стадии являются: специалист DS, в зависимости от квалификации специалиста - на этапе выбора наиболее удачной модели может подлючаться или не подключаться лид DS.

**9.**

Настройка гиперпараметров модели с целью повышения точности модели.
Основными действующими лицами этой стадии являются: специалист DS, в зависимости от квалификации специалиста - может подлючаться или не подключаться лид DS.

**10.**

Оценка экономической и бизнес - выгоды от внедрения в продакшн. При необходимости - А/B/C/etc.-тестирование. Это будет второй ожидаемый результат работы команды.
Основными действующими лицами этой стадии являются: специалист DA, разработчики (на этапе подготовки тестирования), лид DS, специались DS или ML/Data-Ops (для вывода модели в "ограниченный" рамками тестовой группы продакшн).

**11**

В случае наличия экономической и бизнес выгоды - развертывание модели в продакшн (и третий ожидаемый результат работы команды). В случае отсутствия - переход на шаг 12.
Успешное окончания развертывания модели в прод ознаменует окончание проекта "создание модели" и начало продукта "модель" (т.е. в этом случае пункты 12-14 будут выполняться уже в рамках процесса "улучшение модели" в целях продукта "модель").
Основными действующими лицами этой стадии являются: разработчики,  специались DS или ML/Data-Ops (для вывода модели в продакшн и настройки ml-pipline).


**12-13.**

Определение дополнительных необходимых входных данных и возможности их сбора с целью улучшения качества существующей или создания новой модели. Организация сбора и хранение дополнительных входных данных.
Основными действующими лицами этой стадии являются: заказчик, специалисты DS, DA и лид DS - для определения необходимых данных- специалисты BA, DE - для определение возможности сбора данных.

**14.**
Возврат на пункт 2 и повторение всей цепочки с учетом дополнительных входных данных.
