## Решение <a class='anchor' id='3'>

Поскольку подразумевается, что проводилось А/B-тестирование и не мы его начинали, то было бы не плохо, для начала, проверить а верно ли вообще был произведен сбор данных. Правильный сбор данных подразумевает минимум 3 вещи:
- разбиение пользователей - случайное (хочется, чтобы наши группы отличались только воздействием и показатели группы отражали показатели генеральной совокупности)
- у каждого пользователя в течение эксперимента не должна меняться группа (не хочется, чтобы пользователь видел оба варианта, хочется независимости выборок)
- способ разбиения - детерминированный (хотим имет возможность, в случае чего, в точности повторить разбиение, например если захотим проверить взаимодействия этих же изменений с какой-то новой функциональностью, например изменениями на 6 уровне)

Проверить последние 2 пункта не представляется возможным, а вот проверить первый пункт, пускай и не очень точно, но можно - как минимум, количество пользователей должно быть +/- одинаковым в каждой группе. У нас оно как раз +/- одинаковое. Было бы прекрасно еще посмотреть на сырые, а не агрегированные данные, чтобы сказать действительно ли распределение случайно (посмотреть на распределения метрик, они до эксперементов должны быть +/- одинаковыми), но работаем с тем что есть.
 
В обоих случаях "медианое время проведенное пользователем в игре за 1 раз" - больше среднего, при этом средние времена двух вариантов равны (с точностью до 6 десятых секунды), а вот медиана второго варианта больше чем первого. Это говорит о том, что: и в первом и во втором вариантах большая часть пользователей проводит в игре времени меньше чем проводит "средний" пользователь, а показатель среднего времени достигается за счет тех пользователей, которые проводят в игре значительно больше времени по сравнению с основной массой игроков. Тот факт, что во втором варианте медиана все-же ближе к среднему и больше медианы первого варианта, говорит о том, что во втором варианте, большенство пользователей проводят больше времени в игре, по сравнению с первым вариантом. Отсюда можно сделать сразу 2 вывода: 
- если наша цель заключается в том, чтобы пользователи проводили больше времени в игре за 1 заход - нам несомненно нужно выбирать второй вариант;
- нам не следует ориентироваться на показатели "среднего" пользователя. Вместо этого лучше использовать "медианного" пользователя, поскольку он будет лучше отражать характеристики большей части игриков.


В первом и во втором вариантах примерно равное количество пользователей перешли по воронке с этапа регистрации на этап игры. При этом, во втором случае, пользователи в среднем давали меньше ответов чем в первом, но в большем проценте случаев давали эти ответы правильно. Поскольку мы знаем, что во втором варианте большенство пользователей тратило больше времени на игру - можно выдвинуть гипотезу, что задания во втором варианте либо труднее, либо пользователи думают над ними более тщательно. Как и в случае со временем - распределение пользователей по количеству отвеченных вопросов скошено вправо, при этом во втором варианте пользователи отвечают в "медиане" на 1 вопрос больше, возможно именно этим обусловленно то, что пользователи проводят в медиане больше времени в игре за 1 заход.

Касательно стартов рекламмы на пользователя: вполне логично, что у второй группы этих стартов (в среднем) больше - больше времени - больше стартов (помним, что мы ориентируемся на "медианного" пользователя, а не на среднего). Также логично, что даже при увеличении (у второй группы) количества стартов рекламы - доход на пользователя немного, но меньше (нужны сырые данные для оценки стат.значимости). Всему виной - меньшая цена за "милю".

Если мы посмотрим на retention, то увидим, что возвращаемость первого дня изменилась в большую сторону, а вот третьего и седьмого - упала. При этом упал и ретеншен одиннадцатого уровня, а вот шестого - вырос, в то время как третьего - не изменился (конечно опять же, было бы здоров иметь сырые данные, чтобы оценить стат. значимость изменений). Исходя из предположений, что изменения значимы - во втором варианте большее число пользователей достигло шестого уровня и сделало это за меньшее время в днях (ретеншн по дням - падает). 

Итого, что мы имеем:
- Если наша цель заключается в том, чтобы пользователи проводили больше времени в игре за 1 заход - нам несомненно нужно выбирать второй вариант
- Нам не следует ориентироваться на показатели "среднего" пользователя. Вместо этого лучше использовать "медианного" пользователя, поскольку он будет лучше отражать характеристики большей части игриков.
- Во втором варианте пользователи в медиане отвечают на 1 вопрос больше и возможно именно этим обусловлено увеличение "медианного" времени проведенного пользователя в игре.
- Возможно, задания во втором варианте либо труднее, либо пользователи думают над ними более тщательно.
- У второй группы больше стартов рекламы (за счет того, что общее суммарное время игры всех игроков больше), но при этом цена за 1000 показов - меньше, а этом доход на пользователя изменился не значительно. Т.е. во втором варианте мы стали более интересны для рекламодателей (при условии неизменности CTR).
- Во втором варианте большее число пользователей достигло шестого уровня и сделало это на меньшее время, однако, после достежения 6 уровня их интерес к игре стал угасать, как и интерес к игре на 3 день после регистрации. 
- Исходя из всего вышесказанного - возможно что вторым вариантом мы просто изменили ядро нашей аудитории. Следовало бы изучить этот вопрос более тщательно. Создается ощущение, что в первом и втором варианте ожидания пользователей от игры - разные.
- Все выводы не окончательны и могут быть изменены как угодно, если осуществлять проверку на сырых, а не агрегированных данных.

## Формулировка <a class='anchor' id='5'>

В игре внедрили новый функционал: теперь у игроков будут ежедневные миссии (daily missions) и миссии на уровень (level missions). Миссия - это задание, после выполнения которого игрок получает награду.

**Ежедневные миссии**

При заходе в игру в главном меню пользователю показывается поп-ап "задание на сегодня".
Если старт без попапа - то будет иконка на игровом поле и плашки по мере прогресса всплывать будут

**Миссии на уровень.** 

При входе на уровень показывается поп-ап с заданием на уровень.

## Решение <a class='anchor' id='6'>

Исходная гипотеза выдвиннута корректно. Действительно, если мы хотим проверить воздействие двух факторов, то нам нужно производить A/B/C/D тестирование. Однако, прежде чем запускать тест - нужно не только произвессти А/А-тестирование (убедиться, что  в каждой из 4х групп пользователей целевые показатели одинаковы, разбиение пользователей случайно, система сплитования детерменированна, группа, к которой относится пользователе - не изменна на протяжении всего эксперемента), но и сформулировать гипотезы. State of Art подходом является подход, при котором мы за 1 тест проверяется 1 гипотеза. Таким образом, A/B/C/D тест - не что иное, как два A/B теста, которые накладываются друг на друга, а значит, баланс классов (пользователей в группах A и B) в первом тесте (например в определении эффекта от наличия или отсутствия дневной миссии) должен соответствовать балансу классов во втором тесте (в наших предположениях - в тексте на наличие или отсутствие эффекта от миссии на уровень). Таким образом, наши группы должны выглядеть следующим образом:

| **Дневная миссия**/**Уровневая миссия** | **Есть**| **Нет** |      
| -------------- | ------------------------------- |------------------------------- |
| **Есть**           | Есть дневная и уровневая миссия | Есть. только дневная миссия |
| **Нет**            | Есть только уровневая миссия    | Без миссий |

и каждая из 4 групп должна содержать по 25% пользователей отобранных для участия в тестировании.

Вроде бы запущенный тест соответствует этому правилу по верхнему уровню. Но, как известно, дъявол кроется в деталях, а детали таковы, что, если мы проверяем наличие или отсутствие миссии в какой-то день / на каком-то уровне, то это должен быть один и тот же день или один и тот же уровень (по крайней мере, пока мы не провели такие эксперементы и не выявили наилучшие исходы для каждой пары A/B тестов ). В случае же, если мы хотим сравнивать именно то, что мы сравниваем - мы снача должны провести нулевую серию эксперементов, в которых доказать, что каждый из выбранных нами вариантов - лучше альтернативы, а только потом уже сравнивать представленные (в задании) варианты.

К сожалению, в таблице не указан необходимый размер выборки для исследования, поэтому, даже если предположить, что реализация тестирования правильна - мы не можем делать никаких выводов, потому как можем легко попасть в "ошибку подглядывания" и остановить тест раньше (или позже) чем это будет нужно с точки зрения статистики, но так, чтобы результаты подтверждали наших "фаворитов", которые мы осознанно или неосозннно выбрали на этапе подготовки эксперемента. Однако, можно сказать, что имело бы смысл смотреть ряд дополнительных показателей, которые дали бы возможность более целостно оценивать изменения. Например:
- можно смотреть не только медианное, но и среднее время проведенное в игре (мы же в общем хотим его увеличить)
- можно смотреть всевозможные метрики, которые влияют на маркетинговые показатели, такие как LTV, доход с пользователя и прочие (в бизнес-модели чем больше и дольше пользователь играет - тем больше он видит рекламы, а значит, мы получаем больше денег)
- можно смотреть СPC (вполне возможно, что разные комбинации "зайдут" и "не зайдут" разным группам, в зависимости от этого нужно будет смотртеь на разницу профита и потерь и принимать решения на основе того, сколько дополнительной прибыли принисут нам нововведения. Может так быть, что нововведениями мы поменяем ядро нашей ЦА)
